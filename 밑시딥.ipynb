```python
import numpy as np

x = np.array([0, 1])
w = np.array([0.5, 0.5])
b = -0.7

np.sum(x*w) + b #sum()은 입력한 배열에 담긴 모든 원소의 총합 계산

def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(x*w) + b
    if tmp <= 0:
        return 0
    else :
        return 1
    
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(x*w) + b
    if tmp <= 0:
        return 0
    else :
        return 1
    
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.4
    tmp = np.sum(x*w) + b
    if tmp <= 0:
        return 0
    else :
        return 1
    
print(AND(0, 1))
print(NAND(0, 1))
print(OR(0, 1))

def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    return AND(s1, s2)

print(XOR(0, 1))
print(XOR(1, 1))
```

```python
import numpy as np
import matplotlib.pylab as plt

#계단함수 구현하기
def step_function(x):#x는 넘파이 배열
    return np.array(x>0, dtype = np.int) # y를 int로 타입변환하여 return

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) #y축의범위 지정
plt.show
```

```python
import numpy as np
import matplotlib.pylab as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show
```
